# Matplotlib (plot)


```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# =====================================================================
# Matplotlib 기초 핵심 요약 메모
# =====================================================================

# ---------------------------------------------------
# 1. 도화지와 그래프 영역 생성 (기본 구조)
# ---------------------------------------------------
fig, ax = plt.subplots() # fig: 전체 도화지(Figure), ax: 실제 그래프가 그려지는 영역(Axes)


# ---------------------------------------------------
# 2. 데이터 입력 자료형 (Data Types)
# ---------------------------------------------------
# Matplotlib은 '배열 형태'의 데이터라면 대부분 입력받을 수 있습니다.
# (리스트, NumPy 배열, Pandas 시리즈 등을 섞어 써도 동작함)

x_list = [1, 2, 3, 4]                  # 1) 파이썬 기본 리스트
y_np = np.array([1, 4, 2, 3])          # 2) NumPy 배열 (표준 권장)
y_pd = pd.Series([1, 4, 2, 3])         # 3) Pandas 시리즈 (실무에서 자주 쓰임)


# ---------------------------------------------------
# 3. x, y 좌표를 이용한 점/선 그리기 (ax.plot)
# ---------------------------------------------------
# [경우 A] 단일 숫자 입력: 특정 좌표에 '점 1개'만 찍기 (선 안 그려짐)
# ax.plot(1, 3, 'go') # x=1, y=3 위치에 초록색(g) 동그라미(o) 1개

# [경우 B] 배열(리스트 등) 입력: '여러 점'을 찍고 '선'으로 연결하기
ax.plot(x_list, y_np, 'rs--') # 빨간색(r) 네모(s) 마커를 파선/대시선(--)으로 연결


# ---------------------------------------------------
# 4. 스타일 기호 사전 (색상 + 마커 + 선 종류)
# ---------------------------------------------------
# 하나의 문자열 안에 [색상][마커][선종류]를 순서 상관없이 조합 가능!
# 예: 'k:^' (검은색 + 점선 + 삼각형) / 'g-.D' (초록색 + 1점쇄선 + 마름모)

# (1) 색상 (Color)
# r: 빨간색, g: 초록색, b: 파란색, y: 노란색
# k: 검은색 (파란색 b와 구분하기 위해 k 사용), w: 흰색

# (2) 마커 모양 (Marker)
# 'o': 동그라미, 's': 네모, '^': 세모, '*': 별
# 'x': X 표시, 'D': 마름모, '+': + 표시

# (3) 선의 종류 (Line Style)
# '-' : 실선 (기본값)
# '--': 파선/대시선
# '-.': 1점 쇄선
# ':' : 점선


# ---------------------------------------------------
# 5. 그래프 화면에 출력하기
# ---------------------------------------------------
plt.show()
```

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# 1. 두 개의 데이터 선 그리기 (범례를 위해 label 추가!)
x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 2, 4, 1]

# 빨간 점선에는 'Apple'이라는 라벨을, 파란 실선에는 'Banana'라는 라벨을 붙입니다.
ax.plot(x, y1, 'ro--', label='Apple')
ax.plot(x, y2, 'b^-', label='Banana')

# 2. 제목과 축 이름표 달기
ax.set_title("Fruit Sales")       # 그래프 제목
ax.set_xlabel("Month")            # x축 이름
ax.set_ylabel("Sales (kg)")       # y축 이름

# 3. 범례 표시하기 (plot에서 적어둔 label들이 박스로 나타납니다)
ax.legend()

plt.show()
```

```
# 다중 분포 표현하기
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('iris')
df.head(2)

# figure, axes 생성
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(df.index, df['sepal_length'], 'ro', label='sepal') #df['sepal_length]자료형: Series
ax.plot(df.index, df['petal_length'], 'b^', label='petal')

ax.set_xlabel('sepal')
ax.set_ylabel('petal')
ax.set_title('iris')
ax.legend()
plt.show()
```

```
# subplots을 이용하여 2개의 axe 만들기

import matplotlib.pyplot as plt

# 1행 2열(가로로 2칸)로 도화지를 나눕니다.
# 보통 figsize=(가로길이, 세로길이)를 추가해 도화지 크기를 넉넉하게 키워줍니다.
fig, axs = plt.subplots(1, 2, figsize=(10, 4)) # axe가 복수이므로 axs로 표현 #axs는 배열형태가 됨.

# 첫 번째 칸 (왼쪽, axs[0])에 그리기
axs[0].plot([1, 2, 3], [1, 4, 9], 'ro-')
axs[0].set_title("Left Plot")

# 두 번째 칸 (오른쪽, axs[1])에 그리기
axs[1].plot([1, 2, 3], [9, 4, 1], 'bs--')
axs[1].set_title("Right Plot")

plt.show()
```

```
import matplotlib.pyplot as plt

# 1. 2행 2열 도화지 생성
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# ---------------------------------------------------
# ⭐ 전체 도화지(Figure)에 큰 제목 달기
# ---------------------------------------------------
fig.suptitle("My Awesome 4-Panel Chart", fontsize=16, fontweight='bold')

# 2. 개별 그래프 그리기 및 각각의 제목(title) 달기
axs[0, 0].plot([1, 2, 3], [1, 2, 3], 'ro-')
axs[0, 0].set_title("Top Left")

axs[0, 1].plot([1, 2, 3], [3, 2, 1], 'bs--')
axs[0, 1].set_title("Top Right")

axs[1, 0].plot([1, 2, 3], [1, 4, 9], 'g^-')
axs[1, 0].set_title("Bottom Left")

axs[1, 1].plot([1, 2, 3], [9, 4, 1], 'k*:')
axs[1, 1].set_title("Bottom Right")

# 3. 레이아웃 자동 조정
# 전체 제목(suptitle)과 개별 제목(title)이 겹치지 않도록 깔끔하게 정리해 줍니다.
plt.tight_layout()

plt.show()
```

```
# 축의 눈금 설정, lebel 설정 바꾸

import matplotlib.pyplot as plt

# 가로로 2칸짜리 도화지를 만듭니다.
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle("Default vs Customized Limits & Ticks", fontweight='bold')

x = [1, 2, 3, 4, 5]
y = [10, 25, 30, 45, 50]

# ---------------------------------------------------
# [왼쪽 칸] 컴퓨터가 알아서 그린 기본 설정
# ---------------------------------------------------
axs[0].plot(x, y, 'bo-')
axs[0].set_title("1. Default")
# 눈여겨볼 점: y축이 10부터 시작하고 50에서 끝나서 여백이 없고,
# x축 눈금도 1.5, 2.0, 2.5 처럼 소수점으로 지저분하게 나옵니다.

# ---------------------------------------------------
# [오른쪽 칸] 우리가 축 범위와 눈금을 직접 설정한 그래프
# ---------------------------------------------------
axs[1].plot(x, y, 'bo-')
axs[1].set_title("2. Customized")

# 1. y축 범위를 0부터 60까지 넉넉하게 변경
axs[1].set_ylim(0, 60)

# 2. x축 눈금을 우리가 원하는 위치에만 표시하고, 이름도 문자로 변경
axs[1].set_xticks(ticks=[1, 2, 3, 4, 5], labels=['Jan', 'Feb', 'Mar', 'Apr', 'May'])

# 글씨가 겹치지 않게 자동 정리
plt.tight_layout()
plt.show()
```

```
# 표에 화살표 및 주석삽입하기

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 4))

x = [1, 2, 3, 4, 5]
y = [10, 25, 30, 45, 50]
ax.plot(x, y, 'bo-')
ax.set_ylim(0, 60)

# ---------------------------------------------------
# ⭐ 가장 높은 지점에 주석 달기
# ---------------------------------------------------
ax.annotate(
    text='Max Value!',      # 1. 화면에 보여줄 글씨
    xy=(5, 51),             # 2. 화살표 끝이 가리킬 타겟 좌표 (가장 높은 점)
    xytext=(4.2, 56),         # 3. 글씨(Max Value!)를 배치할 시작 좌표 (타겟보다 조금 왼쪽 위에 배치)
    arrowprops=dict(facecolor='yellow', shrink=0.03)  # 4. 화살표 모양 설정 (노란색, 크기 조절)
)

plt.show()
```

```
# (화살표 응용) 다양한 화살표의 표현

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 5))

# 꺾은선 그래프 기본 세팅
x = [1, 2, 3, 4, 5]
y = [10, 30, 20, 50, 40]
ax.plot(x, y, 'ko-', label='Data')
ax.set_ylim(0, 65)

# ---------------------------------------------------
# 1. 뚱뚱한 기본 화살표 (색상 및 투명도 변경)
# ---------------------------------------------------
ax.annotate(
    text='1. Thick & Red',
    xy=(2, 30),
    xytext=(1, 45),
    # 빨간색 면, 파란색 테두리, 반투명(0.5)
    arrowprops=dict(facecolor='red', edgecolor='blue', alpha=0.5, shrink=0.05)
)

# ---------------------------------------------------
# 2. 얇고 날렵한 화살표 (arrowstyle='->' 사용)
# ---------------------------------------------------
ax.annotate(
    text='2. Simple & Green',
    xy=(4, 50),
    xytext=(3, 60),
    # 얇은 선 화살표(->), 초록색, 선 굵기(lw) 2
    arrowprops=dict(arrowstyle='->', color='green', lw=2)
)

# ---------------------------------------------------
# 3. 양방향 화살표 (arrowstyle='<->' 사용)
# ---------------------------------------------------
ax.annotate(
    text='3. Double & Purple',
    xy=(5, 40),
    xytext=(5, 10),
    # 양쪽 화살표(<->), 보라색, 점선(ls='--')
    arrowprops=dict(arrowstyle='<->', color='purple', ls='--')
)

ax.set_title("Various Arrow Styles")
plt.show()
```

```
# twin을 사용하여 쌍둥이 축 만들

import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(7, 4))
fig.suptitle("Shared X-Axis", fontweight='bold')

# ---------------------------------------------------
# 1. 첫 번째 데이터 (왼쪽 y축 사용 - 기온)
# ---------------------------------------------------
days = [1, 2, 3, 4, 5]
temp = [22, 24, 25, 28, 30]

# 빨간색 실선으로 그립니다.
ax1.plot(days, temp, 'ro-')
ax1.set_xlabel("Days")
ax1.set_ylabel("Temperature (°C)", color='red')  # 축 이름도 빨간색으로 맞춤!
ax1.tick_params(axis='y', labelcolor='red')      # 눈금 숫자도 빨간색으로 맞춤!

# ---------------------------------------------------
# 2. 쌍둥이 축 만들기 (오른쪽 y축 생성)
# ---------------------------------------------------
# ax1과 x축을 공유(twin)하는 새로운 축 ax2를 만듭니다.
ax2 = ax1.twinx()

# ---------------------------------------------------
# 3. 두 번째 데이터 (오른쪽 y축 사용 - 판매량)
# ---------------------------------------------------
sales = [1000, 1500, 1200, 4000, 5000]

# 파란색 점선으로 그립니다.
ax2.plot(days, sales, 'b^--')
ax2.set_ylabel("AC Sales (Units)", color='blue') # 축 이름 파란색
ax2.tick_params(axis='y', labelcolor='blue')     # 눈금 숫자 파란색

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np  # 다양한 숫자를 쉽게 만들기 위해 numpy를 잠시 빌려옵니다.

fig, ax = plt.subplots(figsize=(7, 5))

# ---------------------------------------------------
# 1. 흩뿌릴 데이터 준비
# ---------------------------------------------------
# x, y 좌표
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

# 점의 크기 (숫자가 클수록 점이 커짐)
sizes = [100, 300, 500, 700, 900]

# 점의 색상 기준 값 (숫자에 따라 다른 색이 칠해짐)
colors = [10, 20, 30, 40, 50]

# ---------------------------------------------------
# 2. 산점도(Scatter) 그리기
# ---------------------------------------------------
# cmap='viridis'는 파랑-초록-노랑으로 변하는 Matplotlib의 기본 컬러맵입니다.
# (이 외에도 'plasma', 'coolwarm', 'Reds' 등 아주 다양한 테마가 있습니다.)
sc = ax.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.8)

# ---------------------------------------------------
# 3. 컬러바(Colorbar) 달아주기
# ---------------------------------------------------
# 도화지(fig)에 컬러바를 추가합니다. 어떤 색이 어떤 숫자를 의미하는지 알려주는 범례 역할을 합니다.
fig.colorbar(sc, ax=ax, label='Color Intensity')

ax.set_title("Scatter Plot with Colormap")
plt.show()
```
