# Folium 지도 시각화 학습 노트북

Python **folium** 라이브러리로 인터랙티브 지도를 만드는 방법을 단계별로 학습합니다.

| 섹션 | 내용 |
|------|------|
| 0 | 공통 설정 (`show_map` 헬퍼) |
| 1 | 기본 지도 · 마커 |
| 2 | 아이콘 · HTML 팝업 |
| 3 | 여러 마커 · 선 그리기 |
| 4 | 영역 표시 (PolyLine · Polygon · GeoJSON) |
| 5 | 그룹 · 레이어 제어 (FeatureGroup) |
| 6 | 공공데이터 — 전국 무인교통단속카메라 |
| 7 | 히트맵 (정적) |
| 8 | 시간에 따른 히트맵 애니메이션 (HeatMapWithTime) |
| 9 | MiniMap (미니맵) |
| 10 | 거리 · 면적 재기 (MeasureControl) |
| 11 | 마우스 위치 표시 (MousePosition) |
| 12 | 그리기 도구 (Draw) |
| 13 | 세계 국가 지도 (GeoJSON URL) |
| 14 | **시도별 경계 · 카메라 수** (GeoJSON) |

> **각 코드 셀은 독립 실행** — 순서와 관계없이 해당 셀만 실행해도 동작합니다.


## 0. 공통 설정

노트북(Cursor / VS Code)에서 folium 지도를 iframe으로 보여주는 `show_map` 함수를 정의합니다.


```python
import base64
from pathlib import Path

import folium as fm
from IPython.display import IFrame, display


def show_map(m, width=600, height=400, file='map.html'):
    """Cursor/VS Code 노트북에서 folium 지도를 표시."""
    html = m.get_root().render()
    path = Path(file)
    path.write_text(html, encoding='utf-8')
    if len(html) < 500_000:
        encoded = base64.b64encode(html.encode('utf-8')).decode('utf-8')
        display(IFrame(src=f'data:text/html;base64,{encoded}', width=width, height=height))
    else:
        display(IFrame(src=path.name, width=width, height=height))
        print(f'지도 크기 {len(html) / 1024 / 1024:.1f}MB → {path.name} 저장 후 표시')
```

## 5. 그룹 · 레이어 제어 (FeatureGroup)

`FeatureGroup`으로 마커를 그룹화하고, `LayerControl`로 지도 우측 상단에서 그룹별 표시/숨김을 전환합니다.


```python
import base64
from pathlib import Path

import folium as fm
from IPython.display import IFrame, display


def show_map(m, width=600, height=400, file='map.html'):
    """Cursor/VS Code 노트북에서 folium 지도를 표시."""
    html = m.get_root().render()
    path = Path(file)
    path.write_text(html, encoding='utf-8')
    if len(html) < 500_000:
        encoded = base64.b64encode(html.encode('utf-8')).decode('utf-8')
        display(IFrame(src=f'data:text/html;base64,{encoded}', width=width, height=height))
    else:
        display(IFrame(src=path.name, width=width, height=height))
        print(f'지도 크기 {len(html) / 1024 / 1024:.1f}MB → {path.name} 저장 후 표시')

# 5. 그룹 · 레이어 제어 (FeatureGroup)
m = fm.Map(location=[37.4904, 127.0162], zoom_start=15, width=600, height=400)

GROUP_STYLES = {
    '그룹1': {'color': 'red', 'icon': 'home'},
    '그룹2': {'color': 'blue', 'icon': 'star'},
}
groups = {name: fm.FeatureGroup(name=name).add_to(m) for name in GROUP_STYLES}


def add_marker(group_name, location, popup=None, tooltip=None, icon=None):
    style = GROUP_STYLES[group_name]
    fm.Marker(
        location=location,
        popup=popup,
        tooltip=tooltip,
        icon=fm.Icon(color=style['color'], icon=icon or style['icon']),
    ).add_to(groups[group_name])


add_marker('그룹1', [37.4904, 127.0162])
add_marker('그룹1', [37.4937, 127.0132], icon='star')
add_marker('그룹2', [37.4937, 127.0137])
add_marker('그룹1', [37.4900, 127.0162])

fm.LayerControl().add_to(m)
show_map(m)
```

## 6. 공공데이터 활용 — 전국 무인교통단속카메라

[공공데이터포털](https://www.data.go.kr)의 **전국무인교통단속카메라표준데이터** CSV를 사용합니다.

- 파일: `전국무인교통단속카메라표준데이터.csv` (CP949 인코딩)
- `edupython/.env`에 `DATAPORTAL_DECODING` 키가 있으면 API로 최신 데이터를 추가 병합합니다.


```python
# 6.1 데이터 불러오기
import os
import pandas as pd
import requests
from pathlib import Path
from dotenv import load_dotenv
from IPython.display import display

CSV_PATH = '전국무인교통단속카메라표준데이터.csv'
KEY_COL = '무인교통단속카메라관리번호'

mapping_cols = {
    'mnlssRegltCameraManageNo': '무인교통단속카메라관리번호',
    'ctprvnNm': '시도명',
    'signguNm': '시군구명',
    'roadKnd': '도로종류',
    'roadRouteNo': '도로노선번호',
    'roadRouteNm': '도로노선명',
    'roadRouteDrc': '도로노선방향',
    'rdnmadr': '소재지도로명주소',
    'lnmadr': '소재지지번주소',
    'latitude': '위도',
    'longitude': '경도',
    'itlpc': '설치장소',
    'regltSe': '단속구분',
    'lmttVe': '제한속도',
    'regltSctnLcSe': '단속구간위치구분',
    'ovrspdRegltSctnLt': '과속단속구간길이',
    'prtcareaType': '보호구역구분',
    'installationYear': '설치연도',
    'institutionNm': '관리기관명',
    'phoneNumber': '관리기관전화번호',
    'referenceDate': '데이터기준일자',
    'instt_code': '제공기관코드',
}

df_csv = pd.read_csv(CSV_PATH, encoding='CP949')

if load_dotenv(Path.cwd().parent / '.env'):
    dec = os.getenv('DATAPORTAL_DECODING')
    url = 'https://api.data.go.kr/openapi/tn_pubr_public_unmanned_traffic_camera_api'
    params = {'serviceKey': dec, 'pageNo': 1, 'numOfRows': 1000, 'type': 'json', 'installationYear': '2026'}
    response = requests.get(url, params=params)
    response.raise_for_status()
    items = response.json()['response']['body'].get('items')

    if items:
        if isinstance(items, dict):
            items = [items]
        df_api = pd.DataFrame(items).rename(columns=mapping_cols)
        existing_ids = set(df_csv[KEY_COL].astype(str))
        df_new = df_api[~df_api[KEY_COL].astype(str).isin(existing_ids)]
        df = pd.concat([df_csv, df_new], ignore_index=True)
        print(f'CSV {len(df_csv):,}건 + API 신규 {len(df_new):,}건 = 합계 {len(df):,}건')
    else:
        df = df_csv.copy()
        print(f'API 응답 없음. CSV {len(df):,}건만 사용')
else:
    df = df_csv.copy()
    print(f'환경변수 없음. CSV {len(df):,}건만 사용')

print(f'shape: {df.shape}')
print(f'설치연도 범위: {int(df["설치연도"].min())} ~ {int(df["설치연도"].max())}')
display(df.head(2))
```

### 6.2 MarkerCluster — 대량 마커 표시

수만 건의 데이터를 개별 마커로 찍으면 느려지므로 `MarkerCluster`로 묶어 표시합니다.
학습용으로 상위 3,000건만 표시합니다.


```python
import base64
from pathlib import Path

import folium as fm
from IPython.display import IFrame, display


def show_map(m, width=600, height=400, file='map.html'):
    """Cursor/VS Code 노트북에서 folium 지도를 표시."""
    html = m.get_root().render()
    path = Path(file)
    path.write_text(html, encoding='utf-8')
    if len(html) < 500_000:
        encoded = base64.b64encode(html.encode('utf-8')).decode('utf-8')
        display(IFrame(src=f'data:text/html;base64,{encoded}', width=width, height=height))
    else:
        display(IFrame(src=path.name, width=width, height=height))
        print(f'지도 크기 {len(html) / 1024 / 1024:.1f}MB → {path.name} 저장 후 표시')
CSV_PATH = '전국무인교통단속카메라표준데이터.csv'

import pandas as pd

df = pd.read_csv(CSV_PATH, encoding='CP949')

# 6.2 MarkerCluster
from folium.plugins import MarkerCluster


def make_popup_html(row):
    def val(*keys, suffix=''):
        for key in keys:
            v = row.get(key, '')
            if pd.notna(v) and str(v).strip():
                return f'{v}{suffix}'
        return '-'

    rows = [
        ('설치연도', val('설치연도', suffix='년')),
        ('단속구분', val('단속구분')),
        ('제한속도', val('제한속도', suffix=' km/h')),
        ('관리기관', val('관리기관명')),
        ('연락처', val('관리기관전화번호')),
    ]
    table_rows = ''.join(
        f'<tr>'
        f'<td style="padding:6px 10px;color:#64748b;font-size:12px;">{label}</td>'
        f'<td style="padding:6px 10px;color:#1e293b;font-size:13px;font-weight:500;">{value}</td>'
        f'</tr>'
        for label, value in rows
    )
    return f"""
    <div style="font-family:'Malgun Gothic',sans-serif;width:100%;">
        <div style="background:linear-gradient(135deg,#1e3a5f,#2563eb);color:#fff;padding:14px 16px;">
            <div style="font-size:11px;opacity:.85;">무인교통단속카메라</div>
            <div style="font-size:15px;font-weight:700;">{val('설치장소')}</div>
        </div>
        <table style="width:100%;border-collapse:collapse;">{table_rows}</table>
    </div>
    """


SAMPLE_SIZE = 100
m = fm.Map(location=[35.24, 131.10], zoom_start=7, width=1200, height=700)
cluster = MarkerCluster().add_to(m)

for _, row in df.head(SAMPLE_SIZE).iterrows():
    popup = fm.Popup(make_popup_html(row), max_width=320)
    fm.Marker(
        location=[row['위도'], row['경도']],
        popup=popup,
        tooltip=row['설치장소'],
    ).add_to(cluster)

print(f'마커 {SAMPLE_SIZE:,}개 추가 완료')
show_map(m, width=600, height=400, file='map_cluster.html')
```

## 8. 시간에 따른 히트맵 애니메이션 (HeatMapWithTime)

`설치연도`별로 히트맵 프레임을 만들어 **시간 흐름에 따른 설치 확산**을 애니메이션으로 보여줍니다.

| 설정 | 설명 |
|------|------|
| `auto_play=True` | 지도 로드 시 자동 재생 |
| `index` | 각 프레임의 연도 라벨 |
| `radius` | 히트맵 점 크기 |

- **누적 방식** (`설치연도 <= y`): 해당 연도까지 설치된 카메라 전체를 표시 → 확산 추이 파악에 유리
- **연도별 방식** (`설치연도 == y`): 그 해에 새로 설치된 카메라만 표시

아래 코드는 **누적 방식**을 사용합니다. 지도 하단 타임라인 슬라이더로 연도를 직접 조작할 수도 있습니다.


```python
import base64
from pathlib import Path

import folium as fm
from IPython.display import IFrame, display


def show_map(m, width=600, height=400, file='map.html'):
    """Cursor/VS Code 노트북에서 folium 지도를 표시."""
    html = m.get_root().render()
    path = Path(file)
    path.write_text(html, encoding='utf-8')
    if len(html) < 500_000:
        encoded = base64.b64encode(html.encode('utf-8')).decode('utf-8')
        display(IFrame(src=f'data:text/html;base64,{encoded}', width=width, height=height))
    else:
        display(IFrame(src=path.name, width=width, height=height))
        print(f'지도 크기 {len(html) / 1024 / 1024:.1f}MB → {path.name} 저장 후 표시')
CSV_PATH = '전국무인교통단속카메라표준데이터.csv'

import pandas as pd

df = pd.read_csv(CSV_PATH, encoding='CP949')

# 8.1 설치연도별 누적 히트맵 애니메이션
from folium.plugins import HeatMapWithTime

df_time = df.copy()
df_time['설치연도'] = df_time['설치연도'].astype('Int64')

years = sorted(df_time['설치연도'].dropna().unique())
heat_data = []
year_labels = []

for y in years:
    cumulative = df_time[df_time['설치연도'] <= y]
    heat_data.append(cumulative[['위도', '경도']].values.tolist())
    year_labels.append(str(int(y)))

print(f'총 {len(df_time):,}건 → {len(year_labels)}개 프레임 ({year_labels[0]}~{year_labels[-1]})')
for label, data in zip(year_labels[::5], [len(d) for d in heat_data[::5]]):
    print(f'  {label}년 누적: {data:,}건')

m_time = fm.Map(location=[36.24, 128.10], zoom_start=7, width=1200, height=700)
HeatMapWithTime(
    data=heat_data,
    index=year_labels,
    auto_play=True,
    max_opacity=0.8,
    radius=14,
    use_local_extrema=False,
).add_to(m_time)

show_map(m_time, width=1200, height=700, file='map_time.html')
```

### 8.2 (참고) 연도별 스냅샷 방식

그 해에 **새로 설치된** 카메라만 보고 싶다면 아래처럼 `==` 조건을 사용합니다.


```python
import base64
from pathlib import Path

import folium as fm
from IPython.display import IFrame, display


def show_map(m, width=600, height=400, file='map.html'):
    """Cursor/VS Code 노트북에서 folium 지도를 표시."""
    html = m.get_root().render()
    path = Path(file)
    path.write_text(html, encoding='utf-8')
    if len(html) < 500_000:
        encoded = base64.b64encode(html.encode('utf-8')).decode('utf-8')
        display(IFrame(src=f'data:text/html;base64,{encoded}', width=width, height=height))
    else:
        display(IFrame(src=path.name, width=width, height=height))
        print(f'지도 크기 {len(html) / 1024 / 1024:.1f}MB → {path.name} 저장 후 표시')
CSV_PATH = '전국무인교통단속카메라표준데이터.csv'

import pandas as pd

df = pd.read_csv(CSV_PATH, encoding='CP949')

# 8.2 연도별 스냅샷 (그 해 신규 설치만)
from folium.plugins import HeatMapWithTime

df_time = df.copy()
df_time['설치연도'] = df_time['설치연도'].astype('Int64')
years = sorted(df_time['설치연도'].dropna().unique())

heat_data_snap = []
year_labels_snap = []

for y in years:
    yearly = df_time[df_time['설치연도'] == y]
    heat_data_snap.append(yearly[['위도', '경도']].values.tolist())
    year_labels_snap.append(str(int(y)))

print(f'{len(year_labels_snap)}개 프레임 ({year_labels_snap[0]}~{year_labels_snap[-1]})')

m_snap = fm.Map(location=[36.24, 128.10], zoom_start=7, width=1200, height=700)
HeatMapWithTime(
    data=heat_data_snap,
    index=year_labels_snap,
    auto_play=True,
    max_opacity=0.8,
    radius=14,
).add_to(m_snap)

show_map(m_snap, width=1200, height=700, file='map_time_snap.html')
```

## 14. 시도별 경계 · 카메라 수

`SIDO_MAP_2022.json`은 대한민국 **시도 경계 GeoJSON** 파일입니다.
로컬에 없으면 GitHub에서 자동 다운로드됩니다.

- 데이터: [southkorea-maps](https://github.com/southkorea/southkorea-maps)
- CSV `시도명`과 경계를 합쳐 시도별 카메라 수를 툴팁으로 표시


```python
import base64
from pathlib import Path

import folium as fm
from IPython.display import IFrame, display


def show_map(m, width=600, height=400, file='map.html'):
    """Cursor/VS Code 노트북에서 folium 지도를 표시."""
    html = m.get_root().render()
    path = Path(file)
    path.write_text(html, encoding='utf-8')
    if len(html) < 500_000:
        encoded = base64.b64encode(html.encode('utf-8')).decode('utf-8')
        display(IFrame(src=f'data:text/html;base64,{encoded}', width=width, height=height))
    else:
        display(IFrame(src=path.name, width=width, height=height))
        print(f'지도 크기 {len(html) / 1024 / 1024:.1f}MB → {path.name} 저장 후 표시')

# 14. 지역별 데이터 표시 (시도 경계 GeoJSON)
from pathlib import Path

import geopandas as gpd
import pandas as pd
import requests

CSV_PATH = '전국무인교통단속카메라표준데이터.csv'
SIDO_PATH = Path('SIDO_MAP_2022.json')
SIDO_URL = (
    'https://raw.githubusercontent.com/southkorea/southkorea-maps/'
    'master/gadm/json/skorea-provinces-geo.json'
)

NAME_TO_SIDO = {
    'Seoul': '서울특별시',
    'Busan': '부산광역시',
    'Daegu': '대구광역시',
    'Incheon': '인천광역시',
    'Gwangju': '광주광역시',
    'Daejeon': '대전광역시',
    'Ulsan': '울산광역시',
    'Gyeonggi-do': '경기도',
    'Gangwon-do': '강원특별자치도',
    'Chungcheongbuk-do': '충청북도',
    'Chungcheongnam-do': '충청남도',
    'Gyeongsangbuk-do': '경상북도',
    'Gyeongsangnam-do': '경상남도',
    'Jeollabuk-do': '전북특별자치도',
    'Jeollanam-do': '전라남도',
    'Jeju': '제주특별자치도',
}

if not SIDO_PATH.exists():
    print(f'{SIDO_PATH} 없음 → 다운로드 중...')
    response = requests.get(SIDO_URL, timeout=120)
    response.raise_for_status()
    SIDO_PATH.write_bytes(response.content)
    print(f'저장 완료: {SIDO_PATH} ({len(response.content) / 1024 / 1024:.1f}MB)')

data = pd.read_csv(CSV_PATH, encoding='CP949')
sido_counts = data.groupby('시도명').size().reset_index(name='카메라수')

kor = gpd.read_file(SIDO_PATH)
kor['시도명'] = kor['NAME_1'].map(NAME_TO_SIDO)
kor = kor.merge(sido_counts, on='시도명', how='left')
kor['카메라수'] = kor['카메라수'].fillna(0).astype(int)

m = fm.Map(location=[36.24, 128.10], zoom_start=7, width=1200, height=700)

fm.GeoJson(
    kor,
    name='시도 경계',
    style_function=lambda x: {
        'fillColor': '#3b82f6',
        'color': '#1e40af',
        'weight': 1,
        'fillOpacity': 0.4,
    },
    tooltip=fm.GeoJsonTooltip(
        fields=['시도명', '카메라수'],
        aliases=['시도', '카메라 수'],
    ),
).add_to(m)

# 카메라 수에 따라 각 시도별로 색상을 다르게 지정하는 컬러맵 생성
import branca.colormap as cm

max_count = kor['카메라수'].max()
colormap = cm.linear.Blues_09.scale(0, max_count)
colormap.caption = '시도별 무인교통단속카메라 수'

def style_function_color(feature):
    count = feature['properties'].get('카메라수', 0)
    return {
        'fillColor': colormap(count),
        'color': '#1e40af',
        'weight': 1,
        'fillOpacity': 0.6,
    }

fm.GeoJson(
    kor,
    name='시도별 카메라 컬러맵',
    style_function=style_function_color,
    tooltip=fm.GeoJsonTooltip(
        fields=['시도명', '카메라수'],
        aliases=['시도', '카메라 수'],
    ),
).add_to(m)

colormap.add_to(m)

fm.LayerControl().add_to(m)
show_map(m, width=900, height=500, file='map_sido.html')
```
