### 1. 모듈 불러오기 및 도화지 기본 세팅
- `figsize`: 도화지의 크기를 (가로, 세로) 인치 단위로 설정합니다.
- `dpi`: 해상도를 설정합니다. (기본값 보통 100, 고화질은 300 이상)
- `layout='constrained'`: 요소들이 겹치지 않게 여백을 자동 조정하는 최신 옵션입니다. (`tight_layout()`과 유사)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# (선택) 전체 그래프의 스타일 테마를 지정할 수 있습니다.
# 사용 가능한 테마 확인: print(plt.style.available)
# plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots(figsize=(8, 4), dpi=100, layout='constrained')
ax.plot([1, 2, 3], [1, 2, 3])
plt.show()
```

### 2 & 3. 선과 마커의 모든 옵션 (All-in-One)
축약형 기호(`'ro-'`) 대신 명시적인 영단어 옵션을 쓰면 훨씬 세밀한 조작이 가능합니다.

- `color` (또는 `c`): 선 색상. 이니셜(r, g, b, c, m, y, k, w) 뿐만 아니라 Hex 코드(`#FF5733`)도 가능합니다.
- `linestyle` (또는 `ls`): 선 종류 (`'-'`, `'--'`, `'-.'`, `':'`)
- `linewidth` (또는 `lw`): 선의 두께
- `marker`: 점의 모양 (`'o'`, `'s'`, `'^'`, `'*'`, `'D'` 등)
- `markersize` (또는 `ms`): 마커의 크기
- `markerfacecolor` (또는 `mfc`): 마커 내부 색상
- `markeredgecolor` (또는 `mec`): 마커 테두리 색상
- `markeredgewidth` (또는 `mew`): 마커 테두리 두께
- `alpha`: 투명도 (0.0 투명 ~ 1.0 불투명)
- `zorder`: 겹쳤을 때 그리는 순서 (숫자가 클수록 맨 위에 그려짐)

```python
fig, ax = plt.subplots(figsize=(8, 4))

x = [1, 2, 3, 4, 5]
y = [10, 25, 30, 45, 50]

ax.plot(x, y, 
        color='#2CA02C',         # 헥스 코드로 진한 초록색 지정
        linestyle='--',          # 파선(대시선)
        linewidth=2.5,           # 선 두께
        marker='D',              # 마름모 마커
        markersize=12,           # 마커 크기
        markerfacecolor='gold',  # 마커 내부는 금색
        markeredgecolor='red',   # 마커 테두리는 빨간색
        markeredgewidth=2,       # 마커 테두리 두께
        alpha=0.9,               # 약간의 투명도
        zorder=3,                # 맨 앞으로 가져오기
        label='Premium Data')    # 범례 이름

plt.show()
```

### 4. 그래프 꾸미기 끝판왕 (제목, 라벨, 범례, 그리드)
각 글씨들의 크기, 위치, 굵기 등을 상세하게 설정할 수 있습니다.

- **제목**: `loc` (left, center, right), `pad` (그래프와의 간격)
- **라벨**: `labelpad` (축과의 간격)
- **범례 (`ax.legend`)**:
  - `loc`: 'best', 'upper right', 'lower left' 등 내부 위치
  - `bbox_to_anchor`: 튜플을 넣어 범례를 아예 그래프 바깥으로 뺄 때 사용 (예: `(1.05, 1)`)
  - `ncol`: 범례를 몇 개의 열로 나열할지 설정
  - `shadow`: 그림자 효과, `fancybox`: 모서리 둥글게
- **그리드 (`ax.grid`)**: 배경 격자선 (축 방향 지정 가능: `axis='x'`, `axis='y'`, `axis='both'`)

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot([1, 2, 3], [1, 4, 2], label='Data A')
ax.plot([1, 2, 3], [3, 2, 5], label='Data B')

# 1. 제목 상세 설정
ax.set_title("Advanced Styling", loc='left', fontsize=16, fontweight='bold', pad=15, color='darkblue')

# 2. 축 라벨 상세 설정
ax.set_xlabel("X-axis (Time)", fontsize=12, labelpad=10)
ax.set_ylabel("Y-axis (Value)", fontsize=12, labelpad=10, fontstyle='italic')

# 3. 범례 상세 설정
ax.legend(loc='upper right', 
          fontsize=11,
          ncol=2,           # 가로로 2개씩 나열
          shadow=True,      # 그림자
          fancybox=True,    # 둥근 모서리
          framealpha=0.9)   # 범례 배경 투명도

# 4. 격자(Grid) 설정
ax.grid(True, axis='both', color='gray', alpha=0.3, linestyle='-.')

plt.show()
```

### 5. 고급 도화지 분할하기 (축 공유, 비율 조절)
- `sharex=True`, `sharey=True`: 여러 그래프가 동일한 축 범위를 강제로 공유하도록 합니다. (비교하기 좋음)
- `gridspec_kw`: 행이나 열의 크기 비율을 다르게 설정합니다.

```python
# 2행 2열 분할. 가로 비율은 1:2, 세로 비율은 2:1로 다르게 설정
# 또한, 모든 subplot이 x축과 y축의 눈금/범위를 공유함
fig, axs = plt.subplots(2, 2, figsize=(8, 6), 
                        sharex=True, sharey=True, 
                        gridspec_kw={'width_ratios': [1, 2], 'height_ratios': [2, 1]})

fig.suptitle("Advanced Subplots (Shared Axes & Ratios)", fontsize=14, fontweight='bold')

axs[0, 0].plot([1, 2], [1, 5], 'r')
axs[1, 1].plot([1, 2], [3, 2], 'b')

plt.tight_layout()
plt.show()
```

### 6. 축 제어 완벽 가이드 (범위, 눈금, 글씨 회전)
- 데이터 글씨가 길어서 겹칠 때는 `rotation` 옵션으로 각도를 틀어줍니다.
- `ax.invert_yaxis()`: y축의 위아래 숫자를 뒤집을 때 사용합니다.

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot([1, 2, 3, 4], [100, 400, 200, 300], 'ko-')

# 1. 범위 설정
ax.set_ylim(0, 500)

# 2. 눈금 지정 및 이름 설정, 글씨 45도 회전 (긴 단어 겹침 방지)
months = ['January', 'February', 'March', 'April']
ax.set_xticks(ticks=[1, 2, 3, 4], labels=months, rotation=45, ha='right') 
                             # ha='right'는 글자의 오른쪽 끝을 눈금에 맞춘다는 뜻

plt.tight_layout()
plt.show()
```

### 7. 텍스트 및 주석(Annotation) 상세 옵션
- **단순 텍스트 (`ax.text`)**: 화살표 없이 특정 좌표에 글씨만 띄울 때 사용합니다. (`ha`: 수평 정렬, `va`: 수직 정렬)
- **화살표 주석 (`ax.annotate`)**: `arrowprops` 안의 다양한 옵션으로 화살표를 성형합니다.
  - `width`: 화살표 몸통 굵기
  - `headwidth`: 화살표 머리(세모) 너비
  - `headlength`: 화살표 머리 길이
  - `shrink`: 양끝 여백 (0~1 비율)
  - `connectionstyle`: 화살표를 직선이 아닌 곡선으로 휘게 만듭니다. (예: `arc3,rad=0.3`)

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot([1, 2, 3, 4], [10, 30, 20, 40], 'bo-')
ax.set_ylim(0, 50)

# 1. 단순 텍스트 추가 (화살표 없음)
ax.text(x=3, y=15, s="Low Point", fontsize=12, color='red', 
        ha='center', va='top',       # 중앙, 위쪽 기준 정렬
        bbox=dict(boxstyle="round", facecolor='yellow', alpha=0.5)) # 배경 박스 추가

# 2. 디테일한 화살표 주석
ax.annotate(
    "Peak!", 
    xy=(4, 40),               # 타겟 좌표
    xytext=(2.5, 45),         # 텍스트 시작 좌표
    fontsize=14, color='darkgreen', fontweight='bold',
    arrowprops=dict(
        facecolor='green',    # 화살표 색상
        edgecolor='black',    # 테두리 색상
        shrink=0.05,          # 타겟과의 여백
        width=3,              # 몸통 굵기
        headwidth=12,         # 머리 너비
        connectionstyle="arc3,rad=-0.2" # 마이너스(-)를 주면 반대 방향으로 휩니다.
    )
)
plt.show()
```

### 8. 이중 y축 (Twinx) 스케일 매칭 심화
스케일이 다른 두 축을 쓸 때는 색상을 명확히 분리하고, 그리드 선이 겹쳐 지저분해지는 것을 방지해야 합니다.

```python
fig, ax1 = plt.subplots(figsize=(8, 4))
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 2.5, 4, 3.5]
y2 = [1000, 1500, 800, 3000, 2500]

# 첫 번째 축 세팅
ax1.plot(x, y1, 'r-o', lw=2)
ax1.set_ylabel("Rate (%)", color='red', fontweight='bold')
ax1.tick_params(axis='y', labelcolor='red', colors='red') # 틱 색상까지 빨간색으로 동기화
ax1.grid(True, axis='y', linestyle='--', alpha=0.5)

# 두 번째 축 세팅
ax2 = ax1.twinx()
ax2.plot(x, y2, 'b--s', lw=2)
ax2.set_ylabel("Volume (EA)", color='blue', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='blue', colors='blue')
ax2.grid(False) # ax1의 그리드와 겹치면 더러워지므로 보통 하나는 끕니다.

plt.show()
```

### 9. 컬러맵과 산점도 (Scatter 심화)
- `vmin`, `vmax`: 색상 매핑의 최소치와 최대치를 강제로 고정합니다. (데이터 이상치 때문에 색상이 쏠리는 것을 방지)
- `edgecolors`: 각 점의 테두리를 추가해 점들이 겹칠 때 구분을 명확히 합니다.

```python
fig, ax = plt.subplots(figsize=(8, 5))

# 가상의 50개 데이터 랜덤 생성
np.random.seed(42)
x = np.random.rand(50) * 10
y = np.random.rand(50) * 10
sizes = np.random.rand(50) * 800 + 50    # 50 ~ 850 크기
colors = np.random.rand(50) * 100        # 0 ~ 100 값

sc = ax.scatter(x, y, 
                s=sizes, 
                c=colors, 
                cmap='plasma',       # 불타오르는 느낌의 테마
                alpha=0.7,           # 겹치는 부분 색상 혼합을 위한 반투명
                edgecolors='white',  # 점들끼리 겹칠 때 흰색 테두리로 구분
                linewidth=1.5,       # 테두리 두께
                vmin=20, vmax=80)    # 20이하는 가장 어두운색, 80이상은 가장 밝은색으로 고정

cbar = fig.colorbar(sc, ax=ax)
cbar.set_label('Heat Score', rotation=270, labelpad=15, fontweight='bold')

plt.show()
```
